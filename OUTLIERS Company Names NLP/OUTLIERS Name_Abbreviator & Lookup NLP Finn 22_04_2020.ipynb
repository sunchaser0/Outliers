```python
def stringAbbreviator(company_name):
    
    from cleanco import cleanco
#     print(company_name)
    cleanco_results = cleanco(company_name)
#     print(cleanco_results.clean_name())
#     print(cleanco_results.type())
    
    
    cleanco_company_name = str(cleanco_results.clean_name())
    try:
        cleanco_company_type = cleanco_results.type()[0]
    except:
        cleanco_company_type = ""
    #note, if there are multiple company types there needs to be exception handling!
    
#     print("type, name :")
#     print(f"{cleanco_company_type} ... {type(cleanco_company_type)}")
#     print(f"{cleanco_company_name} ... {type(cleanco_company_name)}")
    
    cleaned_names = []
    
    for word in [cleanco_company_name,cleanco_company_type]:
#         print(word)
        word_separated = word.split(" ")[0]
        first_part = word[0:int(len(word_separated)/2)]
        second_part = word[int(len(word_separated)/2):]
        total = first_part + " " + second_part
#         print(total)
#         print("result:")
        cleaned_names.append("".join(letter[0] for letter in total.split()))
    return '-'.join(cleaned_names)

    
```

```python
def getNames():
    
    #This script is to test the name-abbreviator on real companies house data
    
    
    business_names = ["starbucks","mcdonalds","costa","cadbury","inc","MICROSOFT+CORPORATION"]
#     business_names = ["starbucks",]

    starbucks_names = []
    mcdonalds_names = []
    costa_names = []
    cadbury_names = []
    inc_names = []
    microsoft_names = []
    
    for business_name in business_names:
        chosen_business_name = f"https://beta.companieshouse.gov.uk/search?q={business_name}"
        from bs4 import BeautifulSoup
        import requests
        page = requests.get(chosen_business_name)
        soup = BeautifulSoup(page.content)
        
        for tag in soup.find_all('a'):

            for k,v in tag.attrs.items():
        #       
                if v == "View company":
        #     
                    if business_name == "starbucks":
                        starbucks_names.append(tag.text.strip())
                    if business_name == "mcdonalds":
                        mcdonalds_names.append(tag.text.strip())
                    if business_name == "costa":
                        costa_names.append(tag.text.strip())
                    if business_name == "cadbury":
                        cadbury_names.append(tag.text.strip())
                    if business_name == "inc":
                        inc_names.append(tag.text.strip())
                    if business_name == "MICROSOFT+CORPORATION":
                        microsoft_names.append(tag.text.strip())
                    
    #         result.text.strip().translate(str.maketrans('', '', string.punctuation))

    results = starbucks_names + mcdonalds_names + costa_names + cadbury_names + inc_names + microsoft_names
    
    return results
```

```python
def createNameLookup(list_of_company_names):
    
    pairings = []
    new_names = []
    clusters = {}
    
    for name in list_of_company_names:
        counter = 0
        abbreviated_name = stringAbbreviator(name)
        appended_name = abbreviated_name
        while appended_name in new_names:
            counter +=1
            appended_name = f"{abbreviated_name} {counter}"
        new_names.append(appended_name)

    for x in range(0,len(list_of_company_names)):
        pairings.append((list_of_company_names[x],new_names[x]))
        
        
    for string in pairings:
        print(string)

        #can adjust 0:2 to higher, depending on how many entries you have, but needs to be reiterated so that it clusters by
        #character pair, and then by character group

        abbreviation_pair = {}
        if string[1][0:2] not in clusters.keys():
            clusters[string[1][0:2]] = []
        clusters[string[1][0:2]].append(string)
    
    return clusters
    
```

```python
def performNameLookup(name_lookup_dictionary,name_to_match):
    return name_lookup_dictionary[stringAbbreviator(name_to_match)[0:2]]
```
